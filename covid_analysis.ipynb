```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("covid19_Confirmed_dataset.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
df = data.drop(["Lat","Long"], axis = 1 , inplace = True)
```

```python
cda = data.groupby("Country/Region").sum()
```

```python
cda
```

```python
cda.drop(["Province/State"], axis = 1, inplace = True)
```

```python
cda.head()
```

```python
cda.shape
```

```python
cda.loc["China"].plot()
cda.loc["India"].plot()
cda.loc["Spain"].plot()
cda.loc["US"].plot()

plt.legend()
plt.show()
```

```python
cda.loc["China"][:3].plot()
```

```python
cda.loc["India"].diff().plot()
cda.loc["China"].diff().plot()
cda.loc["US"].diff().plot()
plt.legend()
```

```python
countries = cda.index
max_infection_rates = []
for c in countries :
    max_infection_rates.append(cda.loc[c].max())
cda["Max_Infection_Rates"] = max_infection_rates
```

```python
cda
```

```python
corona_data = pd.DataFrame(cda["Max_Infection_Rates"])
```

```python
corona_data
```
